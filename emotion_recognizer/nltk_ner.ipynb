```python
sentence = "James is working at Disney in London."
# 토큰화 후 품사 태깅
tokenized_sentence = nltk.pos_tag(nltk.word_tokenize(sentence))
print(tokenized_sentence)
```

```python
ner_sentence = nltk.ne_chunk(tokenized_sentence)
print(ner_sentence)
```

```python
str = ''
for chunk in ner_sentence:
    print(chunk)
    if hasattr(chunk, 'label'):
        if chunk.label() == 'GPE':
            str += '[LOCATION] '
        else:
            str += '[' + chunk.label() + '] '
    else:
        str += chunk[0] + ' '

print(str)
```

```python
f_ori = open(path + "data/IMSDB/LaLaLand.txt", "r")
```

```python
f_ner = open(path + "data/IMSDB/LaLaLand_ner.txt","w+")
```

```python
while True:
    line = f_ori.readline()
    if not line:
        break
    tokenized_sentence = nltk.pos_tag(nltk.word_tokenize(line))
    ner_sentence = nltk.ne_chunk(tokenized_sentence)
    str = ''
    for chunk in ner_sentence:
        if hasattr(chunk, 'label'):
            if chunk.label() == 'GPE':
                str += '[LOCATION] '
            else:
                str += '[' + chunk.label() + '] '
        else:
            str += chunk[0] + ' '
    f_ner.write(str + '\t0\n')
```

```python
f_ner.close()
f_ori.close()
```
